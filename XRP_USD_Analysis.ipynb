```python
open_close = 'close'
```

```python
xrp_df = pd.read_csv('https://drive.google.com/uc?export=download&id=1-HYltuu-hF7viuxmVOfQS1Q9VK-xmL7g')
xrp_df.head()
```

Clean the data for duplicates, missing values and any formatting errors.

```python
#Convert Unix Dates to DateTime
xrp_df['open time']=pd.to_datetime(xrp_df['open time'], unit='ms')

#drop close time as it is not needed for now
xrp_df.drop('close time', axis=1, inplace=True)
#drop ignore column
xrp_df.drop('ignore', axis=1, inplace=True)

#Set the open time as index
xrp_df.set_index('open time', inplace=True)

#sort by date(index)
xrp_df.sort_index(inplace=True)

xrp_df.head()
```

```python
#See DF info and show Duplicate rows
xrp_df.info()
print("Duplicate rows = ",len(xrp_df[xrp_df.duplicated(keep=False)]))
```

```python
# Show basic info
xrp_df.describe()
```

calculate RSI

```python
#start adding additional Columns for RSI
#Add Change column
xrp_df['change'] = xrp_df['close'].diff(1)
xrp_df.head()
```

```python
#Add gain & loss columns from change
xrp_df['gain'] = xrp_df[xrp_df['change']>0]['change']
xrp_df['gain'] = xrp_df['gain'].fillna(0)

xrp_df['loss'] = xrp_df[xrp_df['change']<0]['change']
xrp_df['loss'] = xrp_df['loss']*-1
xrp_df['loss'] = xrp_df['loss'].fillna(0)

xrp_df.head()
```

```python
#Use python TA to calculate RSI
xrp_df['rsi']=pta.rsi(xrp_df['close'],length = 14)
xrp_df.head(20)
```


What is the trend in price data?

```python
#Plot closing price
xrp_df.plot(y=open_close,kind='line', figsize=(16,8))
plt.xlabel('Date')
plt.ylabel(f'{open_close.capitalize()} Price')
plt.title(f'{open_close.capitalize()} Price Trend')
#plt.yticks([0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9])

#Plot trend for closing price
z = np.polyfit(mdates.date2num(xrp_df.index), xrp_df[open_close],1)
p= np.poly1d(z)

plt.plot(mdates.date2num(xrp_df.index), p(mdates.date2num(xrp_df.index)))

plt.show()
```


What is the volume trend?

```python
#Plot Volume
xrp_df.plot(y='volume',kind='line', figsize=(16,8))
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Volume Trend')

plt.axvspan(mdates.date2num(datetime(2022,2,5)), mdates.date2num(datetime(2022,2,12)),
           label="Mid Q1",color="green", alpha=0.3)
plt.axvspan(mdates.date2num(datetime(2022,5,7)), mdates.date2num(datetime(2022,5,14)),
           label="Mid Q2",color="green", alpha=0.3)
plt.axvspan(mdates.date2num(datetime(2022,8,7)), mdates.date2num(datetime(2022,8,14)),
           label="Mid Q3",color="red", alpha=0.3)
plt.axvspan(mdates.date2num(datetime(2022,9,16)), mdates.date2num(datetime(2022,9,27)),
           label="?",color="green", alpha=0.3)

plt.show()
```


Are there certain days in the month where volume is higher/ lower than normal?

```python
xrp_df['date_day']=xrp_df.index.day
mean = xrp_df['volume'].mean()
median = xrp_df['volume'].median()
volume_avg_by_day = xrp_df.groupby('date_day')['volume'].mean()
volume_avg_by_day.plot(y='volume', figsize=(16,8))
plt.axhline(median)
plt.axhline(mean)
plt.show()
```


Are there certain months where prices are higher / lower than normal?

```python
xrp_df['date_month']=xrp_df.index.month
mean = xrp_df[open_close].mean()
median = xrp_df[open_close].median()
price_avg_by_month = xrp_df.groupby('date_month')[open_close].mean()
price_avg_by_month.plot(y=open_close, kind='bar',figsize=(16,8))
plt.axhline(median)
plt.axhline(mean)
plt.show()
```


Is there a correlation between Volume and Quote Asset Volume?

```python
volume_qav_corr = xrp_df['volume'].corr(xrp_df['quote asset volume'])
#xrp_df.plot.scatter(x = 'volume', y = 'quote asset volume', s = 30, c = 'red', figsize=(16,8));
#plt.rcParams.update({'figure.figsize':(16,8), 'figure.dpi':100})
sns.lmplot(x='volume', y='quote asset volume', data=xrp_df, aspect=2/1, height=7)
plt.text(0.2*(10**9), 1*(10**9), f"Correlation co.={volume_qav_corr}", horizontalalignment='left', size='large', color='green', weight='semibold')
plt.show()
```


Is there a trend in Number of trades? What does it specify?

```python
#Plot Number of trades price
xrp_df.plot(y='number of trades',kind='line', figsize=(16,8))
plt.xlabel('Date')
plt.ylabel('Number of Trades')
plt.title('Number of Trades Trend')

plt.axvspan(mdates.date2num(datetime(2022,2,5)), mdates.date2num(datetime(2022,2,12)),
           label="Mid Q1",color="green", alpha=0.3)
plt.axvspan(mdates.date2num(datetime(2022,5,7)), mdates.date2num(datetime(2022,5,14)),
           label="Mid Q2",color="green", alpha=0.3)
plt.axvspan(mdates.date2num(datetime(2022,8,7)), mdates.date2num(datetime(2022,8,14)),
           label="Mid Q3",color="red", alpha=0.3)
plt.axvspan(mdates.date2num(datetime(2022,9,16)), mdates.date2num(datetime(2022,9,27)),
           label="?",color="green", alpha=0.3)
plt.legend()
plt.show()
```
